```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8, 8))
sns.heatmap(wd_main.corr(),annot=True)
```

```python
from sklearn.ensemble import RandomForestRegressor
train = wd_main[:'2015-12-31']
test = wd_main['2016-01-01':]
```

```python
rfr = RandomForestRegressor()
features = ['PRCP', 'SNOW', 'SNWD', 'TMAX', 'TMIN']
rfr.fit(train[features], train['TempTomorrow'])
```

```python
pred = rfr.predict(test[features])
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(test['TempTomorrow'], pred)
mae
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(train[features], train['TempTomorrow'])
pred2 = lr.predict(test[features])
mae2 = mean_absolute_error(test['TempTomorrow'], pred2)
mae2
```

```python
from sklearn.linear_model import Ridge
rr = Ridge(alpha=0.5)
rr.fit(train[features], train['TempTomorrow'])
pred3 = rr.predict(test[features])
mae3 = mean_absolute_error(test['TempTomorrow'], pred3)
mae3
```

```python
predtable = pd.concat([test['TempTomorrow'], pd.Series(pred3, index = test.index)], axis=1)
predtable.columns = ['Actual', 'Predicted']
predtable

```

```python
predtable.plot(figsize=(20,10))
```
