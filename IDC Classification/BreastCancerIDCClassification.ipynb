```python
d = d.map(lambda a,b: (a/255, b))
sc_i = d.as_numpy_iterator()
batch = sc_i.next()
batch[0].min()
```

```python
fig, ix = plt.subplots(ncols=3, figsize = (20,20))
for idx, img in enumerate(batch[0][:3]):
  ix[idx].imshow(img)
  ix[idx].title.set_text(batch[1][idx])
```

```python
len(d)
```

```python
train = int(len(d)*0.7)
test = int(len(d)*0.1)+1
val = int(len(d)*0.2)+1
```

```python
train
```

```python
test
```

```python
val
```

```python
trainset = d.take(train)
testset = d.skip(train).take(test)
valset = d.skip(train+test).take(val)
```

```python
model = Sequential()
```

```python
model.add(Conv2D(16, (3,3), 1, activation = "relu", input_shape=(256,256,3)))
model.add(MaxPooling2D())

model.add(Conv2D(32, (3,3), 1, activation = "relu", input_shape=(256,256,3)))
model.add(MaxPooling2D())

model.add(Conv2D(16, (3,3), 1, activation = "relu", input_shape=(256,256,3)))
model.add(MaxPooling2D())

model.add(Flatten())

model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile('adam', loss = tf.losses.BinaryCrossentropy(), metrics = ['accuracy'])
```

```python
model.summary()
```

```python
from keras import callbacks
earlystopping = callbacks.EarlyStopping(monitor="val_loss",
                                        mode="min", patience=5,
                                        restore_best_weights=True)

hist = model.fit(trainset, epochs = 15, validation_data = valset, callbacks = [earlystopping])
```

```python
ig = plt.figure()
plt.plot(hist.history['accuracy'], color='blue', label="accuracy")
plt.plot(hist.history['val_accuracy'], color='green', label="validation accuracy")
```

```python
fig = plt.figure()
plt.plot(hist.history['loss'], color='blue', label="loss")
plt.plot(hist.history['val_loss'], color='green', label="validation loss")
```

```python
binaccuracy = BinaryAccuracy()
precision = Precision()
recall = Recall()
for iter in d.as_numpy_iterator():
  X,y = iter
  ypred = model.predict(X)
  binaccuracy.update_state(y,ypred)
  precision.update_state(y,ypred)
  recall.update_state(y,ypred)
```

```python
print(f'Binary Accuracy: {binaccuracy.result().numpy()}')
print(f'Precision: {precision.result().numpy()}')
print(f'Recall: {recall.result().numpy()}')
```
